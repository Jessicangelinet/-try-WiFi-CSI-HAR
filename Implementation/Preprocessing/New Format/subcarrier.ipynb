```python
import pandas as pd

name = 'combined'
session = '2'
type = 'nLoS'
filepath = '../Data/Eye Data/'+f'{type}/'+f'{name}'+f'{session}'+'_eye.csv'

df = pd.read_csv(filepath, sep=',')
display(df)
```

```python
# null_pilot_subcarrier = [0, 1, 2, 3, 63, 62, 61, 32, 53, 39, 11, 25]
null_pilot_subcarrier = [0, 1, 2, 3, 31, 60, 61, 62, 63]

columns_remove = [f'amplitude_{i}' for i in null_pilot_subcarrier] + [f'phase_{i}' for i in null_pilot_subcarrier] + ['timestamp'] + ['activity']

df_cleared = df.drop(columns=columns_remove)

display(df_cleared)
```

## Data for Preprocessing (No column names)

```python
save_filepath = '../Data/subcarrier/'+f'{type}/'+f'{name}'+f'{session}'+'_subcarrier.csv'

display(df_cleared)

df_cleared.to_csv(save_filepath, index=False, header=False)
```

## Has column names

```python
save_path = '../Data/subcarrier/'+f'{type}/header/'+f'{name}'+f'{session}'+'_header.csv'
df_cleared.to_csv(save_path, index=False)
```
