## Adds column names from header.csv file

```python
import pandas as pd

name = 'combined'
session = '2'
type = 'nLoS'
file1 = '../Data/Wireless Eye Data/Preprocessed Data/'+f'{type}/'+f'{name}'+f'{session}'+'_preprocessed.csv'
file2 = '../Data/subcarrier/'+f'{type}/header/'+f'{name}'+f'{session}'+'_header.csv'

data1 = pd.read_csv(file1, header=None)
data2 = pd.read_csv(file2)

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
df1.columns = df2.columns

display(df1)
```

## Adds 'timestamp' and 'activity' column

```python
file3 = '../Data/Eye Data/'+f'{type}/'+f'{name}'+f'{session}'+'_eye.csv'

data3 = pd.read_csv(file3)

df3 = pd.DataFrame(data3)

df1['timestamp'] = df3['timestamp']
timestamp_column = df1.pop('timestamp')  # Remove 'timestamp' column
df1.insert(0, 'timestamp', timestamp_column)
df1['activity'] = df3['activity']
# df1 = df1.drop(columns=['amplitude_31'])
display(df1)
```

```python
save_path = '../Data/Wireless Eye Data/Training Data/'+f'{type}/'+f'{name}'+f'{session}'+'_training.csv'
df1.to_csv(save_path, index=False)
```
